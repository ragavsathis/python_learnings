# Illustrating the importance of threading with basic examples from Corey Schafer Video

## Ex- 1 : Basic problem statement which needs to Optimised 

```python
import time

start_time = time.perf_counter()

def sleep_for_1second():
    print("Going to sleep for 1 seconds !!!!\n")
    time.sleep(1)
    print("*** Finished sleeping ***\n")

sleep_for_1second()
sleep_for_1second()

end_time = time.perf_counter()
print(f"Finished in {round(end_time - start_time,2)} seconds")
```

## Using threading to understand the importance of thread in IO cound tasks

```python
import threading

start_time = time.perf_counter()

t1 = threading.Thread(target=sleep_for_1second)
t2 = threading.Thread(target=sleep_for_1second)
t1.start()
t2.start()
t1.join()
t2.join()

end_time = time.perf_counter()
print(f"Finished in {round(end_time - start_time,2)} seconds")
```

```python
start_time = time.perf_counter()

thread_list = list()
for _ in range(100):
    t = threading.Thread(target=sleep_for_1second)
    t.start()
    thread_list.append(t)
    
for thread in thread_list:
    thread.join()
    
end_time = time.perf_counter()
print(f"Finished in {round(end_time - start_time,2)} seconds")
```
